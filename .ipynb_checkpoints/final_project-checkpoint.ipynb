# Batting average

```python
# average by year groups a dataframe by some input variables, indexes them
# and return a new dataframe

def average_by_year(dataframe, group_column, column1, column2, index):
    group_column_col1_col2 = dataframe[[group_column, column1, column2]]
    sum_grouped = group_column_col1_col2.groupby(group_column).sum()
    if index == 'post':
        key = 'avg_post'
    elif index == 'regular':
        key = 'avg_regular'
    else:
        key = 'avg'
        
    sum_grouped[key] = sum_grouped[column1]/sum_grouped[column2]
    
    return sum_grouped
```

```python
batting_average = average_by_year(batting_df,'yearID','H','AB','regular')
```

```python
batting_post_average = average_by_year(batting_post_df, 'yearID', 'H', 'AB','post')
```

```python
# rename some column names to avoid same column names
batting_average = batting_average.rename(columns = {'AB':'AB_regular'})
batting_post_average = batting_post_average.rename(columns = {'AB':'AB_post'})
```

```python
batting_average.head(2)
```

```python
batting_post_average.head(2)
```

```python
# plot simplifies plotting required graphes

%matplotlib inline
def plot(dataframe1, dataframe2, plot_column_df1, plot_column_df2, xlabel = '', ylabel = '', log_y=False, log_x=False):
    new_dataframe = pd.concat([dataframe1[plot_column_df1], dataframe2[plot_column_df2]], axis=1)
    new_dataframe.plot(figsize=(15,5), logy=log_y, logx=log_x)
    plt.ylabel(ylabel)
    plt.xlabel(xlabel)
    new_dataframe.plot(kind='box', figsize=(4,5), logy=log_y, logx=log_x)
    plt.ylabel(ylabel)
```

```python
plot(batting_average, batting_post_average, 'avg_regular', 'avg_post', xlabel = 'year', ylabel = 'batting average')
```

```python
plot(batting_average, batting_post_average, 'AB_regular', 'AB_post', xlabel = 'year', ylabel = 'at bats', log_y=True)
```

# 2. comparison batting averages (regular vs. post) of post season players

```python
# sum_grouped sum-groupes a dataframe by some input keys
def sum_grouped(dataframe, key1, key2):
    sum_grouped = dataframe.groupby([key1, key2], as_index=False).sum()
    return sum_grouped
```

```python
batting_reduced = batting_df[['yearID', 'playerID', 'H', 'AB']]
batting_reduced.tail(2)
```

```python
batting_post_reduced = batting_post_df[['yearID', 'playerID', 'H', 'AB']]
batting_post_reduced.tail(2)
```

```python
batting_post_reduced_grouped = sum_grouped(batting_post_reduced, 'yearID', 'playerID')
batting_post_reduced_grouped.tail(10)
```

```python
batting_reduced_grouped = sum_grouped(batting_reduced, 'yearID', 'playerID')
batting_reduced_grouped.tail(5)
```

```python
batting_reduced_psp = pd.merge(batting_reduced_grouped, batting_post_reduced_grouped, how='inner', on=['yearID', 'playerID'])
batting_reduced_psp.tail(10)
```

```python
batting_by_years_psp = batting_reduced_psp.groupby('yearID').sum()
```

```python
# rename column names in order to avoid mistakes
batting_by_years_psp = batting_by_years_psp.rename(columns = {'H_x':'H_regular_psp', 'AB_x':'AB_regular_psp', 'H_y':'H_post', 'AB_y':'AB_post'})
batting_by_years_psp.tail(5)
```

```python
# creating new columns using vector algebra in order to determine averages 
batting_by_years_psp['avg_regular_psp']= batting_by_years_psp['H_regular_psp']/batting_by_years_psp['AB_regular_psp']
batting_by_years_psp['avg_post']= batting_by_years_psp['H_post']/batting_by_years_psp['AB_post']
batting_by_years_psp.tail(5)
```

```python
# plot2 just simplifies plotting graphes

def plot2(dataframe, y1, y2, xlabel='', ylabel=''):
    new_dataframe = dataframe[[y1, y2]]
    new_dataframe.plot(figsize=(15,5))
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    new_dataframe.plot(kind='box', figsize=(4,5))
    plt.ylabel(ylabel)
```

```python
plot2(batting_by_years_psp,'avg_regular_psp','avg_post', xlabel='year', ylabel='batting average')
```

```python
plot(batting_average, batting_by_years_psp, 'avg_regular', 'avg_regular_psp', xlabel='year', ylabel='batting average')
```

```python
batting_by_years_psp.describe()
```

## Correlation batting average - salary average

### Salary

```python
# determining relative deviation of salary average
relative_salary = pd.concat([average_salary_by_year, average_salary_psp], axis=1)
relative_salary['relative_deviation_salary'] = (relative_salary['average_salary_psp']-relative_salary['average_salary'])/relative_salary['average_salary']
```

```python
relative_salary.head(2)
```

```python
relative_salary['relative_deviation_salary'].plot(style='r')
plt.xlabel('year')
plt.ylabel('relative deviation salary average')
```

### Batting

```python
# determining relative deviation of batting average
relative_batting = pd.concat([batting_average, batting_by_years_psp], axis=1, join='inner')
relative_batting['relative_deviation_batting'] = (relative_batting['avg_regular_psp']-relative_batting['avg_regular'])/relative_batting['avg_regular']
relative_batting.head(2)
```

```python
relative_batting['relative_deviation_batting'].plot()
plt.xlabel('year')
plt.ylabel('relative deviation batting average')
```

## relative salary deviaton vs. relative batting deviation

```python
# creating a new dataframe containing all datas since 1985
relative = pd.concat([relative_salary, relative_batting], axis=1, join='inner')
relative.head()
```

```python
relative['relative_deviation_salary'].describe()
```

### Chi-Square Test

```python
stats.chisquare(relative['avg_regular'], relative['avg_regular_psp'])
```

```python
stats.chisquare(relative['average_salary'], relative['average_salary_psp'])
```

```python
stats.chisquare(relative['relative_deviation_salary'], relative['relative_deviation_batting'])
```

### Correlation

```python
# creating a new dataframe in order to plot a scatterplot
correlation_1 = pd.DataFrame()
correlation_1['relative_deviation_batting'] = relative['relative_deviation_batting']
correlation_1['relative_deviation_salary'] = relative['relative_deviation_salary']
```

```python
correlation_1.corr(method='pearson', min_periods=1)
```

```python
correlation_1.plot(x='relative_deviation_batting', y='relative_deviation_salary', kind='scatter', color='r')
plt.xlabel('relative deviation batting average')
plt.ylabel('relative deviation salary average')
```

```python
# plotting a jointplot in order to see the correlation
g = sns.jointplot("relative_deviation_batting", "relative_deviation_salary", data=correlation_1, kind='reg', color="r", size=7)
plt.xlabel('year')
plt.ylabel('relative deviation salary average')
```
